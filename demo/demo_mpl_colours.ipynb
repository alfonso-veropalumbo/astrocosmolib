```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set two quantities
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x) + np.cos(x)
```

```python
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.plot(x, y3, label='sin(x)+cos(x)')

plt.xlabel("x")
plt.xlabel("y")
plt.legend()
plt.title("Before changing colors")
```

```python
change_mpl_default_colours()

plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.plot(x, y3, label='sin(x)+cos(x)')

plt.xlabel("x")
plt.xlabel("y")
plt.legend()
plt.title("After changing colors")
```
